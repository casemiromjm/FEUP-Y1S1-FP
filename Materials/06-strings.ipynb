# 6 Data types: Strings

## 6.1 A compound data type

- So far we have seen built-in types like `int`, `float`, `bool`,
    `str` and we've briefly lists and pairs.

- Strings, lists, and pairs are qualitatively different from the
    others because they are made up of smaller pieces.

- In the case of strings, they're made up of smaller strings each
    containing one **character** in a particular order from left to
    right.

- Types that comprise smaller pieces are called **collection** or
    **compound data types**.

- Depending on what we are doing, we may want to treat a compound data
    type as a single thing.

- A string that contains no characters, often referred to as the **empty string**, is still considered to be a string

### Text sequence type — `str`

  - Textual data in Python is handled with `str` objects, or strings.
  
  - Strings are immutable sequences of Unicode code points.
  
  - String literals are written in a variety of ways:

    - Single quotes: `'allows embedded "double" quotes'`;

    - Double quotes: `"allows embedded 'single' quotes"`;

    - Triple quoted: `'''Three single quotes'''`, `"""Three double quotes"""`.

  - Triple quoted strings may span multiple lines - all associated whitespace will be included in the string literal.

    - used in *docstrings* and comments.

https://docs.python.org/3/library/stdtypes.html#textseq

## 6.5 Traversal and the `for` loop

- A lot of computations involve processing a string *one character at
    a time*.

- Often they start at the beginning, select each character in turn, do
    something to it, and continue until the end.

- This pattern of processing is called a **traversal**.

```python
  word = "Banana"
  for letter in word:
      print(letter)
```

$\Rightarrow$
<https://github.com/fp-leic/public/tree/master/lectures/06/straversal.py>

Since a string is simply a sequence of characters, the `for` loop *iterates* over each character automatically.

````python
words = "Monty Python"
for l in words:
    print(l)
````

Recall that the loop variable takes on each value in the sequence of names. The body is performed once for each name.

````python
for aname in ["Eric", "Graham", "John", "Michael", "Terry"]:
    invitation = "Hi " + aname + ".  Please come to my party on Saturday!"
    print(invitation)
````

### Traversal and looping: by Index

- It is also possible to use the `range()` function to systematically generate the indices of the characters.

- The `for` loop can then be used to iterate over these positions.

- These positions can be used together with the indexing operator to access the individual characters in the string.

- This approach is useful when you need to access more than a single character in the loop.

Try this (and then see it using [Python tutor](http://www.pythontutor.com/visualize.html#mode=edit)):

````python
fruit = "apple"
for idx in range(5):
    current_char = fruit[idx]
    print(current_char)
````

Likewise:


````python
fruit = "apple"

position = 0
while position < len(fruit):
    print(fruit[position])
    position = position + 1
````

## 6.10 The `find()` function

### The `enumerate()` function

- `enumerate()` is a built-in function of Python.

```python
 enumerate(iterable, start=0)
```

- `enumerate()` pairs each element of the sequence with its order count.

- Returns an *enumerate object*.

- The `iterable` argument must be a sequence (e.g., a string), an *iterator*, or some other object which supports iteration.

We can consume the result of `enumerate` with a `for` statement:

````python
quote = "Always look"
for counter, char in enumerate(quote):
    print(counter, char)
````

Try the following examples:

````python
some_list = ['apple', 'banana', 'grapes', 'pear']
for counter, value in enumerate(some_list):
    print(counter, value)
````

````python
for counter, value in enumerate(some_list, 1):
    print(counter, value)
````

### A `find()` function

- In a sense, finding an element is the opposite of the indexing operator.

- What does the following function do?

```python
def my_find(haystack, needle):
    """ No clues here! """
    for index, letter in enumerate(haystack):
        if letter == needle:
            return index
    return -1
```

Compare the output of the code above with what Python does itself:

````python
def my_find(haystack, needle):
  for index, letter in enumerate(haystack):
      if letter == needle:
          return index     # eureka traversal or short-circuit evaluation
  return -1

haystack = "Bananarama!"
print("Mine:", my_find(haystack,'a'))
print("Python's:", haystack.find('a'))
````

## 6.12 Optional parameters

- Let's say we want to find the locations of the occurrence of a character in a string starting at a given position.

- We can modify the `find()` function as a `find_2()` function, adding a third parameter for the starting position in the search string.

- Better still, we can combine `find()` and `find_2()` using an **optional parameter**, if we assume 0 as the default starting position:

```python
  def find_2(haystack, needle, start=0):
      for index,letter in enumerate(haystack[start:]):
          if letter == needle:
              return index + start
      return -1
```

$\Rightarrow$
<https://github.com/fp-leic/public/tree/master/lectures/06/optional.py>

Let's do it:

````python
def find_2(haystack, needle, start):
    for index,letter in enumerate(haystack[start:]):
        if letter == needle:
            return index + start
    return -1

print(find_2("banana", "a", 2) == 3)
````

Another way of doing it with a `while` loop:

````python
def find_3(astring, needle, start=0):
    """
    Find and return the index of a char (the needle) in a string (astring).
    Return -1 if the needle does not occur in astring.
    """
    ix = start
    while ix < len(astring):
        if astring[ix] == needle:
            return ix
        else:
            ix = ix + 1
    return -1

print(find_3('banana', 'a', 2))
````

## 6.16 The string `format()` method

- The easiest and most powerful way to format a string in Python3 is
    to use the `format()` method.

- The template string contains place holders, `"... {0} ... {1} ... {2} ..."`.

- The `format()` method substitutes its arguments into the place holders.

- To see how this works, let's start with a few examples:

```python
  phrase = "His name is {0}!".format("Arthur")
  print(phrase)

```

### Format specification

- Each of the replacement fields can also contain a **format specification**.

- This modifies how the substitutions are made into the template, and can control things like:

    - Whether the field is aligned to the left `<`, center `^`, or right `>`;

    - The width allocated to the field within the result string (a number like 10);

    - The type of conversion (we'll initially only force conversion to float, `f` or perhaps we'll ask integer numbers to be converted to hexadecimal using `x`);

    - If the type conversion is a float, you can also specify how many decimal places are wanted (typically, `.2f` is useful for working with currencies to two decimal places).

## 6.17 f-strings

From Python 3.6 onwards there are also *f-strings*, an alternative way to format strings:

```python
name = "Eric"
age = 74
f"Hello, {name}. You are {age}."
```

Because f-strings are evaluated at runtime, you can put any and all valid Python expressions in them:

````python
f"{2 * 37}"
````

But you could also call functions. Here’s an example:

````python
def to_lowercase(input):
    return input.lower()

name = "Eric Idle"
f"{to_lowercase(name)} is funny."
````

You also have the option of calling a method directly:

````python
f"{name.lower()} is funny."
````

*f-strings* are faster than both *%-formatting* (not explained here) and `str.format()` (explained above).

As you already saw, *f-strings* are expressions evaluated at runtime rather than constant values.
